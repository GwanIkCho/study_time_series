```python
import FinanceDataReader as fdr
import pandas as pd

# 종목 코드와 종목 이름을 매칭하는 딕셔너리
symbol_name_dict = {
    '005930': 'sam',
    '068270': 'sel'
}

# 가져올 종목 코드 리스트
columns = list(symbol_name_dict.keys())

# FinanceDataReader로 데이터 가져오기
f_df = fdr.DataReader(columns, start='2010-01-01')

# 종목 코드를 종목 이름으로 변경
f_df.columns = [symbol_name_dict[col] for col in f_df.columns]

# 데이터프레임 확인
print(f_df)
```

```python
f_df.info()
```

```python
import matplotlib.pyplot as plt

f_df.plot(figsize=(12, 8), subplots=True)
plt.show()
```

```python
f_df.diff().plot(figsize=(12, 8), subplots=True)
plt.show()
```

```python
f_df.pct_change().mean().plot(kind='bar', figsize=(7, 5))
```

```python
import numpy as np

rate_f_df = np.log(f_df / f_df.shift(1))
rate_f_df
```

```python
rate_f_df[['sam', 'sel']].plot(figsize=(10, 5), lw=0.5)
```

```python
# 연율화
# 연간 영업일(약 252일로 계산)
rate_f_df.mean() * 252
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
rate_f_df = rate_f_df.dropna()
get_vif(rate_f_df)
```

```python
rate_f_df.hist(figsize=(5, 5), bins=50)
plt.show()
```

```python
# 일간 수익률
rate_f_df.cumsum().apply(np.exp).plot(figsize=(12, 6))
plt.show()
```

```python
# 월간 수익률
rate_f_df.cumsum().apply(np.exp).resample('1m').last().plot(figsize=(12, 6))
plt.show()
```

```python
sam_df = f_df[['sam']].dropna()
sam_df
```

```python
window = 20

sam_df['min'] = sam_df['sam'].rolling(window=window).min()
sam_df['mean'] = sam_df['sam'].rolling(window=window).mean()
sam_df['std'] = sam_df['sam'].rolling(window=window).std()
sam_df['median'] = sam_df['sam'].rolling(window=window).median()
sam_df['max'] = sam_df['sam'].rolling(window=window).max()

sam_df.dropna()
```

```python
import matplotlib.pyplot as plt

ax = sam_df[['min', 'mean', 'max']].iloc[-252:].plot(figsize=(12, 6), style=['g--', 'r--', 'g--'], lw=0.8)
sam_df['sam'].iloc[-252:].plot(ax=ax)
plt.title("AMZN 20-Day Moving Average Price Movement")
plt.show()
```

```python
# SMA(Simple Moving Average): 일정 기간동안의 가격의 평균을 나타내는 보조지표
# 1달 영업일을 21일로 가정, 1년 영업일을 252일로 가정

sam_df['SMA1'] = sam_df['sam'].rolling(window=21).mean() #short-term
sam_df['SMA2'] = sam_df['sam'].rolling(window=252).mean() #long-term
sam_df[['sam', 'SMA1', 'SMA2']].tail()
```

```python
# 아마존 주가 기술 분석
# 골든 크로스, 데드 크로스
sam_df.dropna(inplace=True)

sam_df['positions'] = np.where(sam_df['SMA1'] > sam_df['SMA2'], 1, -1)  # 1: buy , -1: sell /

ax = sam_df[['sam', 'SMA1', 'SMA2', 'positions']].plot(figsize=(15, 6), secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((-0.05, 1))

plt.title("AMZN Trading Window based on Technical Analysis")
plt.show()
```
